```python
import ast
#CHANGE GENRES to only name of genres and remove id
#ast.literal_eval change strings into list of dictionary because we need list to loop over to take only name and also to  merege all this column we need it in list form not strings
#it will convert only a single row to apply on all row we have to use applu function
def convert(obj):
  lst=[]
  for item  in ast.literal_eval(obj):
    lst.append(item['name'])
  return lst
movies.iloc[0].genres
```

```python
movies['genres'] =   movies['genres'].apply(convert)
```

```python
#change movie keyword to only have name that is tag like space travel or romance
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.iloc[0].cast
# here we will extract only top 3 actor names from a movie
def convert1(obj):
  lst=[]
  a=0
  for i in ast.literal_eval(obj):
    if(a<3):
       lst.append(i['name'])
       a=a+1
    else:
      break
  return lst
movies['cast']=movies['cast'].apply(convert1)
```

```python
movies.head()
```

```python
movies.iloc[0].crew
```

```python
#convert into list of ooonly directors name
def convert_crew(obj):
  lst=[]
  for i in ast.literal_eval(obj):
    if(i['job']=='Director'):
      lst.append(i['name'])
  return lst
```

```python
movies['crew']=movies['crew'].apply(convert_crew)
```

```python
movies.head()
```

```python
#make overview to a list so we can process it
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
# now we will remove space from names between first and last name to remove ambiguit like if one name is sam and other last name is sam than it will get confuse to get which one
# also it will make fanstasy horror which a genre but it is look like two different genere so we will remove space btw them
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
#make a new colmn tags so well we use that column to find similar movies maching all tags values
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
movies['tags'][0]
```

```python
#new dataframe with less and important features
new_mov = movies[['movie_id','title','tags','runtime','year']]
```

```python
#make tags list to string
new_mov['tags'] = new_mov['tags'].apply(lambda x:" ".join(x))
print(new_mov.head())
new_mov['tags'] = new_mov['tags'].apply(lambda x: x.lower())
print(new_mov.iloc[0].tags)
new_mov.isnull().sum()
```

```python
print(new_mov['tags'][0])
```

```python
# in tags there is problem that words like action and actions have same meaning
#so we will use stemming to make it same so our syystem do not get confused and treat them as different entitity
#so there is a nltk libarry to do this
from nltk.stem.porter import PorterStemmer
pt = PorterStemmer()
def stemi(text):
  p=[]
  for i in  text.split():
    p.append(pt.stem(i))
  return " ".join(p)
#lets check with a exmple
stemi('In the 22nd century, a paraplegic Marine is dispatched to the moon Pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. Action Adventure Fantasy ScienceFiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d SamWorthington ZoeSaldana SigourneyWeaver JamesCamero')
#now we will do this for all tag rows
new_mov['tags'] = new_mov['tags'].apply(stemi)
new_mov.head()
new_mov['tags'].shape
```

```python
 #now we will remove stop words like is, a, are bcs tha is not gonna impact our decion making
#than we will create matrix to store all word pointing to an number and if we got a new movie than we check distance of that  movie
#matrix from other movie matrix  (i.e, distance between vectors of one movie and other based on features )
#less distnce show more similarity

# lets have egmaple with TfidfVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

documents = [
    "Action thriller adventure",
    "Romantic comedy drama",
    "Science fiction space opera"
]

vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(documents)
print(X)
feature_names = vectorizer.get_feature_names_out()
print("Feature Names:")
print(feature_names)

# lets have egmaple with  CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(max_features=1000,stop_words='english')
X = vectorizer.fit_transform(documents).toarray()
print(X)

#TfidfVectorizer gives more foucus on unique values and less weight to coomon values
#  CountVectorizer  gives values to words
#TF-IDF(t,d,D)=TF(t,d)×IDF(t,D)
#TF(t,d)= Number of times term t appears in document d / total number of terms in document d
```

```python
#now we will do transformation from string to numercal array on our movie data
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')
vector = vectorizer.fit_transform(new_mov['tags']).toarray()
print(vector)
feature_names = vectorizer.get_feature_names_out()
print(feature_names)
print((vector.shape))

array2 = np.array(new_mov['year'])
print(array2.shape)
#reshape
array2 = array2.reshape(-1, 1)
# Append Series as a new row
vector = np.concatenate((vector, array2), axis=1)
```

WE will use cosine distance instead of eculidian distance  bcs In high-dimensional spaces, the distances between any two points tend to become very similar. This is because the relative difference between the distances of points becomes less significant as the number of dimensions increases. cosine and similarity inversely proportional

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
# function gets movie name that of user will want similar movies
def movie_recom(movie):
  # first it will match movie name with data and give index or row no. of taht movie
  movie_index = new_mov[new_mov['title'] == movie].index[0]
  #than we will get distances to all other movies from taht movie
  #and least distance willl be most similar so we have to sort
  #but if we sort than index will get diturbed any index can be at any postion
  # than we canniot see this index distrance is related to which movie  so
  #we have to make something that will have relation to every movie index that will not change
  #so enumrate function will do that
  distances = similarity[movie_index]
  movie_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:20]
  #lamda help to sort value based on distances and not on enumurate given index
  #[1:20] will make sure it will only give 20-1 = 19 movie reccomendtion
  # i[0] will give movie row no. realted to distance and iloc will takes us to that column tahn we will use itle to get movie titles
  for i in movie_list:
    print(new_mov.iloc[i[0]].title)

```
